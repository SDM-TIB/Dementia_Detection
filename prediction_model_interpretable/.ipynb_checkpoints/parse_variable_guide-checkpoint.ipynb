```python
import pandas as pd
import numpy as np
import pickle
import collections
```

```python
pd.options.display.max_columns=999
pd.options.display.max_colwidth=-1
```

```python
df = pd.read_excel('../../../data_off/optima/optima_data/Variable_Guide.xlsx')
df.head(5)
```

```python
df['Sub Category '] = df['Sub Category '].fillna(method='ffill')
df.head(5)
```

```python
df.columns.values[0:10]
```

```python
def find_categorical_variable(text):
    #print (text, type(text))
    if type(text)!=float:
        if 'code-decode' in text:
            return True
        else:
            return False
    else:
        return False

df['categorical_variable'] = [find_categorical_variable(desc) for desc in df['Description'].values.tolist()]
```

```python
df['Variables'] = df['Sub Category '] +': '+ df['Variable Label'].apply(lambda x: x[2:])
```

```python
cols  = df.columns.tolist()
cols = cols[0:4]+cols[-2:]+cols[4:-2]
df = df[cols]
df.head(5)
```

```python
df_cat = df[df['categorical_variable']==True]
df_cat_var = df_cat.iloc[:,5:]
df_cat_var.head(5)
```

```python
%%time

def extract_cat(str_cat_num):
    try:
        num = None
        cat = None
        for i, k in enumerate(str_cat_num.split('=')):
            k = k.strip()
            if i == 0:
                num = int(k)
            else:
                cat = str(k)       
        return (num,cat.lower().capitalize())
    except:
        return (None, None)
    
def row_value(row):
    row = row.dropna()
    # print (dict([extract_cat(i) for i in row.iloc[1:].values]))
    return dict(sorted(dict([extract_cat(i) for i in row.iloc[1:].values]).items()))

dict_variable_cat = {row['Variables'].strip():row_value(row) for index, row in df_cat_var.iterrows()}
```

```python
with open("data_varaible_cat.pkl", "wb") as f:
    pickle.dump(dict_variable_cat, f)

with open("data_varaible_cat.pkl", "rb") as f:
    output = pickle.load(f)
    print(output)
```
